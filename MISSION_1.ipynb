 que l'indexing parcour le tableau soit par indice ou par listes d'indices ou même par intervalles
 2. slice c'est une vue et on peut le verifier avec array.flags.owndata si True c'est une copie sinon une vue
 3. comme erreurs courantes avec les mask booleens nous avons 

```python
        # - Cas pratique
table = np.arange(0,1000)
mask = table%7==0
print("les multiples de 7 dans table sont:", table[mask])
def func(x):
    resp = True
    if x==0:
        resp = False
    for i in range(2,x):
        if x%i==0:
            resp = False
            break
    return resp
vf = np.vectorize(func)
mask = vf(table)==True
print("\nles nombres premiers dans table sont:", table[mask])
```

              C. Broadcasting
 -QRO:
 1. si on fait 100x1+1x100 il sera créé un tableau de dimension 100x100 rempli par la somme element pa element de chaque tableau
 2. comme règles du broadcasting nous avons:
     <br>-si les dimmensions sont differentes mes compatibles numpy aligne à droite et ajoute des dimensions de taille 1 à gauche si necessaire
     <br>-deux dimensions sont egales si: elles sont egales ou si l'une d'elle est 1
     <br>-si les dimensions ne sont pas compatibles une erreur est levé    

```python
A = rng.uniform(0,100,(50,50))
v = np.arange(50)
print(f"A + v =\n {A+v}\n")
print(f"chaque colonnes multiplier par son indice+1 donne:\n {A*(v+1)}")
print(f"La moyenne par colonne donne:\n {A.mean(axis=0)}\n")
print(f"le centre de chaque colonne donne:\n {(A[0,:]+A[49,:])/2}")
```

                       D. Strides et mémoire
        -QRO:
1. arr.strides pour arr=np.arange(12).reshape(3,4) donne (32,8)
2. arr.T changes les strides car la transposé inverse les axes par consequent la disposition du tableau et iverse aussi les strides
3. oui<br>
       -Cas Pratique:

```python
arr = np.arange(1_000_000).reshape(1000,1000)
print(f"les strides de arr: {arr.strides}\n les stride de arr.T: {arr.T.strides}")
print("arr.T est une copi:", arr.T.flags.owndata is True,"\n arr.T est une vue:", arr.T.flags.owndata is False)
arr1 = arr[::2,::2]
arr1[0,0] = 25
print("arr1:\n",arr1,"\n arr:\n",arr)
```

```python
                            # E. Mission Data Science réelle
dt = np.dtype([
    ('id','i4'),
    ('age','i4'),
    ('weight','f4'),
    ('height','f4'),
    ('cholesterol','f4'),
    ('blood_pressure','f4'),
    ('BIM','f4')
    
])
rng = np.random.default_rng(42)
dat = np.zeros(5000,dtype=dt)
dat['id'] = np.arange(5000)
dat['age'] = rng.integers(18,90,5000)
dat['weight'] = rng.uniform(50,120,5000)
dat['height'] = rng.uniform(150,200,5000)
dat['cholesterol'] = rng.uniform(120,300,5000)
dat['blood_pressure'] = rng.uniform(80,180,5000)
dat['BIM'] = dat['weight']/((dat['height']/100)**2)
np.save('datset.npy',dat)
```

```python
data = np.load('dataset.npy')
mask = ((data['BIM']>25)&(data['cholesterol']>200))
print("BIM>25 & cholesterol>200: \n", data[mask])
mask = ((data['blood_pressure']>140)&(data['BIM']>30))
print("blood_pressure>140 | BIM>30: \n", data[mask])
                    # details ou stats pour rapport
mask_surpoids = ((25<data['BIM'])&(data['BIM']<29.9))
mask_cholestrolup = data['cholesterol']>=240
mask_hyper_obesité = ((140<data['blood_pressure'])|(30<data['BIM']))
P_surpoids = mask_surpoids.mean()*100
P_cholestrolup = mask_cholestrolup.mean()*100
P_hyper_obesité = mask_hyper_obesité.mean()*100
mask_age_1 = data['age']<30
mask_age_2 = ((data['age']>=30)&(data['age']<55))
mask_age_3 = data['age']>=55
moyenne_tranche_age_1 = data['BIM'][mask_age_1].mean()
moyenne_tranche_age_2 = data['BIM'][mask_age_2].mean()
moyenne_tranche_age_3 = data['BIM'][mask_age_3].mean()
print(f"Pourcentage de surpoids: {P_surpoids:.2f}%\n Pourcentage de cholesterol élevé: {P_cholestrolup:.2f}%\n Pourcentage d'hypertension ou obésité: {P_hyper_obesité:.2f}%\n Moyenne d'IMC pour la tranche d'âge de 18-29: {moyenne_tranche_age_1:.2f}\n Moyenne d'IMC pour la tranche d'âge de 30-54: {moyenne_tranche_age_2:.2f}\n Moyenne d'IMC pour la tranche d'âge de 55-90: {moyenne_tranche_age_3:.2f}\n")
```
